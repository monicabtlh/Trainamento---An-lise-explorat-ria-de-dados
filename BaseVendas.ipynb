## Análise do faturamento no período

```python
total_vendas = df['valor'].sum()
print(f"Faturamento total do período: R$ {total_vendas:,.2f}")
```

```python
# =MAIOR no Excel → .max() no pandas
# =MENOR no Excel → .min() no pandas

#Primeiro, é necessário  a conversão do dado para o formato "data"
df['data'] = pd.to_datetime(df['data'], dayfirst=True)

data_inicial = df['data'].min()
data_final = df['data'].max()

print(f"Período analisado: {data_inicial} até {data_final}")
```
